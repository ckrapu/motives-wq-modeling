### Linear model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=827)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred_train = lr_model.predict(X_train)
y_pred_test = lr_model.predict(X_test)

# Print smallest / largest coefficients as well as 90% range
coefs = lr_model.coef_
print("Smallest coefficients:")
print(coefs[np.argsort(coefs)[:10]])
print("Largest coefficients:")
print(coefs[np.argsort(coefs)[-10:]])
print("90% range of coefficients:")
print(np.percentile(coefs, 90) - np.percentile(coefs, 10))


r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

print(f"Linear Regression R² on training set: {r2_train:.4f}")
print(f"Linear Regression R² on test set: {r2_test:.4f}")
```

```python
import xgboost as xgb
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt

dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=pretty_predictor_cols.tolist())
dtest = xgb.DMatrix(X_test, label=y_test, feature_names=pretty_predictor_cols.tolist())

params = {
    'objective': 'reg:squarederror',
    'max_depth': 6,
    'eta': 0.1,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 827  
}

num_boost_round = 100
xgb_model = xgb.train(params, dtrain, num_boost_round)

y_pred_xgb_train = xgb_model.predict(dtrain)
y_pred_xgb_test = xgb_model.predict(dtest)

r2_xgb_train = r2_score(y_train, y_pred_xgb_train)
r2_xgb_test = r2_score(y_test, y_pred_xgb_test)

print(f"XGBoost R² on training set: {r2_xgb_train:.4f}")
print(f"XGBoost R² on test set: {r2_xgb_test:.4f}")
```

### Declare model

```python
USE_GAM          = False
RANDOM_SUBSAMPLE = True
USE_GP           = False
USE_DAG          = False
TRACK_MU         = True
MULTILEVEL_BETA  = True
ADD_INTERCEPT    = True
FLOAT_FORMAT = "float32"

n = len(y)
p = X.shape[1]

pytensor.config.floatX = FLOAT_FORMAT

y = y.astype(FLOAT_FORMAT)
X = X.astype(FLOAT_FORMAT)
W = W.astype(FLOAT_FORMAT)
coords = coords.astype(FLOAT_FORMAT)

if ADD_INTERCEPT:

    # Check to make sure no constant column is present
    if np.all(np.abs(X.mean(axis=0)) < 1e-6):
        X = np.concatenate([np.ones((X.shape[0], 1), dtype=FLOAT_FORMAT), X], axis=1)
        p = X.shape[1]
        pretty_predictor_cols = np.concatenate([["intercept"], pretty_predictor_cols], axis=0)


if USE_GAM:
    spline_df = 3
    X_splines = np.stack([patsy.bs(x, df=spline_df) for x in X.T], axis=-1)
    dims['spline_degree']=np.arange(spline_df)

print(f"Preparing to run model inference with {X.shape[0]} samples and {X.shape[1]} predictors")
print(f"Range of response values is {y.min()} to {y.max()}")
print(f"Range of predictor values is {X.min()} to {X.max()}")

if RANDOM_SUBSAMPLE:
    is_used = np.random.rand(X.shape[0]) < 0.2
    n = is_used.sum()
else:
    is_used = np.ones(X.shape[0], dtype=bool)
    n = len(y)

dims = {'predictor': pretty_predictor_cols, "obs_id": np.arange(n)}


# If the multilevel beta option is used, the coefficients
# are given a multilevel prior in which the huc12 coefficients are grouped
# around a huc8 mean, which is in turn grouped around a huc4 mean
# and those are grouped around a huc2 mean. A global mean is also
# present at the top level. Each covariate gets a different scale parameter
# at each level of the model, the with the scale prior parameter getting
# progressively smaller from top to bottom. The integer-coded huc12, huc8, etc
# are stored as huc12_int and so on in final_gdf.
final_gdf_subset = final_gdf.loc[is_used].copy()
with pm.Model(coords=dims) as sparse_graph_model:

    X_data = pm.Data('X_data', X[is_used], dims=['obs_id', 'predictor'])

    if MULTILEVEL_BETA:
        unique_per_level = [1] + [len(final_gdf_subset[f'huc{i}'].unique()) for i in [2, 4, 8]]

        n_unique_combined = sum(unique_per_level)
        n_levels = len(unique_per_level)

        cumulative_unique = np.cumsum(unique_per_level)

        # Convention on levels is that 0 is for global mean, 1 is for huc2, 2 is for huc4, and 3 is for huc8.
        level_as_int = np.zeros(n_unique_combined, dtype=int)
        ptr = 0

        for i, n_codes in enumerate(unique_per_level):
            level_as_int[ptr:ptr+n_codes] = i 
            ptr += n_codes

        # Make exactly one row is marked as `0`
        assert np.sum(level_as_int == 0) == 1

        # We use a decreasing sequence of scale parameters
        # to encourage greater and greater shrinkage as we go down the levels
        beta_scales = pm.HalfNormal('beta_scales', sigma=np.asarray([5.0, 1.0, 0.2, 0.04]), shape = [p, n_levels])
        print(f"Found {n_unique_combined} unique huc2, huc4, huc8 values. Breakdown: {unique_per_level}")

        # Create random variables with N(0,1) prior. These will be rescaled and shifted
        # to create the final coefficients later on
        z = pm.Normal('z', mu=0, sigma=1, shape = [n_unique_combined, p])
        
        # Rescale the z by picking off the scale parameters for each level
        #[n_unique_combined, p] * [n_unique_combined, p]
        z_scaled = z * beta_scales[:, level_as_int].T

        # Next, our strategy is to add the 0 level to the 1 level, then add the 1 level to the 2 level, and so on
        # until we get to the huc8 level. We do this by iteratively indexing using the codes from final_gdf_subset
        # and then summing the results. The final result is a [n, p] array of coefficients
        beta = pm.math.zeros((n, p))

        
        # Add global effect (row 0 of z_scaled)
        # Use slicing [0:1, :] to maintain 2D shape for broadcasting safety, although [0, :] usually works
        beta += z_scaled[0:1, :]

        # Add HUC2 effects (indices 1 to cumulative_unique[1]-1)
        start = cumulative_unique[0] # = 1
        end = cumulative_unique[1]   # = 1 + num_unique_huc2
        z_scaled_block = z_scaled[start:end]
        # Ensure codes are numpy array of integers for indexing
        codes = final_gdf_subset['huc2'].astype('category').cat.codes.values
        beta += z_scaled_block[codes]

        # Add HUC4 effects (indices cumulative_unique[1] to cumulative_unique[2]-1)
        start = cumulative_unique[1]
        end = cumulative_unique[2]
        z_scaled_block = z_scaled[start:end]
        codes = final_gdf_subset['huc4'].astype('category').cat.codes.values
        beta += z_scaled_block[codes]

        # Add HUC8 effects (indices cumulative_unique[2] to cumulative_unique[3]-1, which is n_unique_combined-1)
        start = cumulative_unique[2]
        end = cumulative_unique[3] # This equals n_unique_combined
        z_scaled_block = z_scaled[start:end]
        codes = final_gdf_subset['huc8'].astype('category').cat.codes.values
        beta += z_scaled_block[codes]

        beta = pm.Deterministic('beta', beta, dims=['obs_id', 'predictor'])
        
        mu = pm.math.sum(beta * X_data, axis=1)

    else:    
        intercept = pm.Normal('intercept', mu=y.mean(), sigma=y.std() * 2)
        beta_sd = pm.HalfNormal('beta_sd', sigma=5)
        beta = pm.Normal('beta', mu=0, sigma=beta_sd, dims='predictor')
        mu = intercept + X[is_used] @ beta

    # Spatial random effect using geographic coordinates
    if USE_GP:
        ell = pm.Beta('ell', alpha=2, beta=2)
        eta = pm.HalfNormal('eta', sigma=0.05) # Push it to be closer to zero
        cov_func = eta**2 * pm.gp.cov.Matern52(input_dim=2, ls=ell)
        gp = pm.gp.HSGP(m=[50, 50], c=1.5, parametrization= "centered", cov_func=cov_func)
        eps = y[is_used] - mu
        f = gp.prior("f", X=coords[is_used], hsgp_coeffs_dims="basis_coeffs", gp_dims="obs_id")
        mu += f
        
       
    # Storing `mu` can take a lot of memory; this controls
    # whether or not it is stored in the trace
    if TRACK_MU:
        pm.Deterministic('mu', mu)


    # This part creates a likelihood for y ~ N(mu, (I-W)ᵀΩ(I-W)) where W is the adjacency matrix
    # and Ω is a diagonal matrix with entries ω
    # This is a Gaussian DAG model with a sparse precision matrix
    if USE_DAG:
        ε = pm.math.constant(y) - mu
        ω = pm.HalfNormal('ω', 5) # controls the diagonal entries of the precision matrix
        γ = pm.HalfNormal("γ", 5) # controls the off-diagonal entries of the precision matrix
        G_pt = sparse.as_sparse_or_tensor_variable(W[is_used][:, is_used])

        # Make ε be shape (n,1)
        ε_col = ε[:, None]  

        # Do all the ops as (n,1) => (n,1)
        ε_minus_γG_ε  = ε_col - sparse.structured_dot(G_pt,  γ * ε_col)
        ε_minus_γGT_ε = ε_col - sparse.structured_dot(G_pt.T, γ * ε_col)

        # Convert back to (n,) before summing
        resid1 = ε_minus_γG_ε.ravel()
        resid2 = ε_minus_γGT_ε.ravel()

        # For quadratic form εᵀ(I-γG)ᵀΩ(I-γG)ε
        # This is equivalent to ω * (ε - γGε)(ε - γGε)ᵀ
        quadratic_form = ω * pt.sum(resid1 * resid2)
            
        logdet = n * pt.log(ω)
        logp =  -0.5 * (n * pt.log(2 * np.pi) + quadratic_form) + 0.5 * logdet
        
    else:
        sigma = pm.HalfCauchy('sigma', beta=1)
        likelihood = pm.Normal('likelihood', mu=mu, sigma=sigma, observed=y[is_used])
```

### Bayesian regression $R^2$ calculation

```python


ylims = (y.min(), y.max())
xlims = (y.min(), y.max())

# Calculate R² for the full model
y_pred = mu + f
r_squared = 1 - np.var(y - y_pred) / np.var(y)
print(f"Full model R²: {r_squared:.4f}")

# Calculate predictions without spatial effect
# Extract the spatial effect from the trace
spatial_effect = trace.posterior.f.mean(dim=['chain', 'draw']).values

# Subtract spatial effect from full predictions to get predictions without spatial component
y_pred_no_spatial = mu
r_squared_no_spatial = 1 - np.var(y - y_pred_no_spatial) / np.var(y)
print(f"R² without spatial effect: {r_squared_no_spatial:.4f}")

# Create figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot full model predictions
axes[0].scatter(y, y_pred, alpha=0.5, s=10)
axes[0].plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Ideal Fit')
axes[0].set_xlabel('True Values')
axes[0].set_ylabel('Predicted Values')
axes[0].set_title(f'Full Model Predictions\nR² = {r_squared:.4f}')
axes[0].legend()
axes[0].set_xlim(xlims)
axes[0].set_ylim(ylims)

# Plot predictions without spatial effect
axes[1].scatter(y, y_pred_no_spatial, alpha=0.5, s=10)
axes[1].plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Ideal Fit')
axes[1].set_xlabel('True Values')
axes[1].set_ylabel('Predicted Values')
axes[1].set_title(f'Predictions Without Spatial Effect\nR² = {r_squared_no_spatial:.4f}')
axes[1].legend()
axes[1].set_xlim(xlims)
axes[1].set_ylim(ylims)

plt.tight_layout()
plt.savefig(VISUALS_DIR / 'prediction_comparison.png', dpi=300)
plt.show()

# Calculate the proportion of variance explained by spatial effect
spatial_contribution = r_squared - r_squared_no_spatial
print(f"Proportion of variance explained by spatial effect: {spatial_contribution:.4f} ({spatial_contribution/r_squared:.1%} of total R²)")
```

```python
import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS
from numpyro.handlers import scope, plate, seed
from numpyro import deterministic, sample

import jax
import jax.numpy as jnp
# Assuming W is a scipy.sparse.csr_matrix
# Make sure scipy is installed: pip install scipy
import scipy.sparse as sp # Keep for potential input type checking or manipulation

# --- HSGP Squared Exponential Kernel Functions (2D Implementation) ---

def spectral_density_se(w, alpha, length):
    """
    Spectral density for the Squared Exponential kernel (up to constant factors).
    Matches the form used in the NumPyro example 'hsgp.py'.
    Note: This is often proportional to sqrt(S(w)) where S is the power spectral density.
    """
    # alpha: GP amplitude (std dev)
    # length: GP length scale
    # w: Frequency (sqrt of eigenvalue)
    c = alpha * jnp.sqrt(2 * jnp.pi) * length # Constant factor differs across definitions
    e = jnp.exp(-0.5 * (length**2) * (w**2))
    return c * e

def get_se_eigenpairs_2d(L, M_per_dim):
    """
    Calculate the first M=M_per_dim*M_per_dim eigenvalues and
    related frequency components (omega) for the Laplacian on [-L, L] x [-L, L].
    """
    # Indices for basis functions in each dimension
    m_vals = jnp.arange(1, M_per_dim + 1) # m1, m2 indices from 1 to M_per_dim

    # Create 2D grid of indices
    m1, m2 = jnp.meshgrid(m_vals, m_vals, indexing='ij')
    m1 = m1.flatten() # Shape (M_total,)
    m2 = m2.flatten() # Shape (M_total,)

    # Calculate frequency components (omega)
    omega_m1 = m1 * jnp.pi / (2 * L)
    omega_m2 = m2 * jnp.pi / (2 * L)

    # Calculate 2D eigenvalues
    # lambda = omega1^2 + omega2^2
    eigenvalues_2d = omega_m1**2 + omega_m2**2

    return eigenvalues_2d, omega_m1, omega_m2, m1, m2

def diag_spectral_density_se_2d(alpha, length, L, M_per_dim):
    """
    Computes the diagonal elements of the spectral density matrix (sqrt)
    for the Squared Exponential kernel HSGP approximation in 2D.
    Essentially sqrt(S(sqrt(lambda_m))) where lambda_m are the 2D eigenvalues.
    """
    eigenvalues_2d, _, _, _, _ = get_se_eigenpairs_2d(L, M_per_dim)
    # Frequencies are sqrt of eigenvalues
    w = jnp.sqrt(eigenvalues_2d)
    # Calculate spectral density values using the 1D SE spectral density function
    # applied to these 2D frequencies
    spd_values = spectral_density_se(w, alpha, length)
    return spd_values # Shape (M_total,)

def eigenfunctions_se_2d(coords, L, M_per_dim):
    """
    Computes the first M=M_per_dim*M_per_dim eigenfunctions for the SE kernel
    approximation via Laplacian eigenfunctions on [-L, L] x [-L, L], evaluated at `coords`.
    phi_{m1,m2}(x,y) = sin(omega_m1*(x+L)) * sin(omega_m2*(y+L)) / L (normalization factor)
    """
    n = coords.shape[0]
    M_total = M_per_dim * M_per_dim

    _, omega_m1, omega_m2, _, _ = get_se_eigenpairs_2d(L, M_per_dim) # Shapes (M_total,)

    # Expand dims for broadcasting: coords (n, 2), omegas (M_total,)
    x = coords[:, 0:1] # Shape (n, 1)
    y = coords[:, 1:2] # Shape (n, 1)
    omega_m1 = omega_m1[None, :] # Shape (1, M_total)
    omega_m2 = omega_m2[None, :] # Shape (1, M_total)

    # Calculate basis functions: sin(omega * (coord + L))
    term1 = jnp.sin(omega_m1 * (x + L)) # Shape (n, M_total) via broadcasting
    term2 = jnp.sin(omega_m2 * (y + L)) # Shape (n, M_total) via broadcasting

    # Combine and normalize
    # Normalization factor for 2D eigenfunctions on [-L,L]x[-L,L] is 1/L
    phi = (term1 * term2) / L # Shape (n, M_total)

    return phi


# --- Hilbert Space Approximate GP (SE Kernel, 2D) Component ---
def approx_se_gp_ncp(coords, L, M_per_dim, name_prefix="gp"):
    """
    Hilbert space approximation for the Squared Exponential kernel GP (2D coords)
    in the non-centered parametrisation.

    Args:
        coords (jnp.ndarray): Coordinates array, shape (n, 2). Assumed scaled appropriately for L.
        L (float): Boundary parameter for the Hilbert space [-L, L] x [-L, L].
        M_per_dim (int): Number of basis functions per dimension. Total M = M_per_dim * M_per_dim.
        name_prefix (str): Prefix for NumPyro sample sites.

    Returns:
        jnp.ndarray: The latent GP function values, shape (n,).
    """
    M_total = M_per_dim * M_per_dim

    # Priors for GP hyperparameters
    alpha = sample(f"{name_prefix}_alpha", dist.HalfCauchy(1.0))
    length = sample(f"{name_prefix}_length", dist.InverseGamma(2.0, 0.5)) # Adjust prior as needed

    # Get eigenfunctions and spectral density
    phi = eigenfunctions_se_2d(coords, L, M_per_dim) # Shape (n, M_total)
    spd_values = diag_spectral_density_se_2d(alpha, length, L, M_per_dim) # Shape (M_total,)

    # Sample basis coefficients (non-centered)
    with plate(f"{name_prefix}_basis", M_total):
        beta_gp = sample(f"{name_prefix}_beta", dist.Normal(0, 1)) # Shape (M_total,)

    # Compute GP realization
    # Ensure spd_values has the correct shape for broadcasting if needed, though it should be (M_total,)
    f_gp = deterministic(f"{name_prefix}_f", phi @ (spd_values * beta_gp)) # Shape (n,)
    return f_gp


# --- Main Spatial Regression Model ---
def spatial_regression_hsgp_car(
    X, coords, W_sym, L, M_per_dim=10, y=None, use_sparse_car=True
):
    """
    NumPyro model for spatial regression using HSGP (SE Kernel) and CAR random effects.

    Args:
        X (jnp.ndarray): Predictor matrix, shape (n, p). Assumed scaled.
        coords (jnp.ndarray): Coordinates matrix, shape (n, 2). Assumed scaled if needed by HSGP funcs.
        W_sym (jnp.ndarray or scipy.sparse matrix): Symmetric adjacency matrix.
        L (float): HSGP boundary parameter for [-L, L] x [-L, L].
        M_per_dim (int): Number of basis functions per dimension for HSGP.
        y (jnp.ndarray, optional): Response variable, shape (n,). Assumed scaled. Defaults to None.
        use_sparse_car (bool): If True and W_sym is sparse, use sparse CAR implementation.
    """
    n, p = X.shape
    assert coords.shape[0] == n
    assert W_sym.shape == (n, n)
    assert coords.shape[1] == 2 # Ensure coords are 2D

    # --- Regression Coefficients ---
    beta = sample("beta", dist.Normal(0., 30.).expand([p])) # Shape (p,)

    # --- Intercept ---
    intercept = sample("intercept", dist.Normal(0., 100.))

    # --- HSGP Spatial Random Effect (Squared Exponential Kernel, 2D) ---
    gp_effect = approx_se_gp_ncp(coords, L=L, M_per_dim=M_per_dim, name_prefix="gp") # Shape (n,)

    # --- CAR Network Random Effect ---
    car_rho = sample("car_rho", dist.Uniform(0, 1)) # Spatial correlation often positive
    car_tau = sample("car_tau", dist.HalfCauchy(1.)) # Conditional precision

    car_dist = dist.CAR(
        loc=0.0,
        correlation=car_rho,
        conditional_precision=car_tau,
        adj_matrix=W_sym,
        is_sparse=use_sparse_car,
    )
    car_effect = sample("car_effect", car_dist) # Shape (n,)

    # --- Linear Predictor ---
    mu = deterministic("mu", intercept + X @ beta + gp_effect + car_effect) # Shape (n,)

    # --- Likelihood ---
    sigma = sample("sigma", dist.HalfCauchy(1.)) # Observation noise scale

    with plate("data", n):
        sample("obs", dist.Normal(mu, sigma), obs=y)
key = jax.random.PRNGKey(1) # Use different seed
key_x, key_coords, key_w, key_beta, key_gp, key_car, key_noise, key_mcmc = jax.random.split(key, 8)


print("Setting up MCMC...")
L_hsgp = 1.2 # Boundary slightly larger than 1.0
M_per_dim_hsgp = 10

kernel = NUTS(spatial_regression_hsgp_car, init_strategy=numpyro.infer.init_to_value(values={'car_rho': 0.5}))
mcmc = MCMC(
    kernel,
    num_warmup=500,
    num_samples=500, # Increase samples for real analysis
    num_chains=1,    # Increase chains for convergence checks (requires numpyro.set_host_device_count)
    progress_bar=True
)

# 3. Run MCMC
print("Running MCMC...")
#numpyro.set_host_device_count(1) # Set this if num_chains > 1
mcmc.run(
    key_mcmc,
    X=X,
    coords=coords, # Pass scaled coords
    W_sym=W_sym,
    L=L_hsgp,
    M_per_dim=M_per_dim_hsgp,
    y=y, # Pass scaled y
    use_sparse_car=True
)

# 4. Analyze results
print("\nMCMC Summary:")
mcmc.print_summary()
samples = mcmc.get_samples()

```
